# # JSON Client Examples

```python
import http.client
import json

conn6 = http.client.HTTPSConnection('www.httpbin.org')

headers6 = {'Content-type': 'application/json'}

foo = {'text': 'Hello HTTP #1',
      'jenny': 8675309,
      'myfloat': 3.1415}

json_data = json.dumps(foo)

conn6.request('POST', '/post', json_data, headers6)

response6 = conn6.getresponse();
body_byte = response6.read();
body_str = body_byte.decode();
```

```python
pprint(response6.getheaders())
```

```python
print(f'{type(body_str)}')
print(body_str)
# pp(body_str)
```

```python
j_data = json.loads(body_str);
pprint(j_data)
```

### https://realpython.com/python-json/

```python
import json
import requests

response = requests.get("https://jsonplaceholder.typicode.com/todos")
todos = json.loads(response.text);
```

```python
# response is case-insensitive dict, so convert it to a regular dict for pprint
pprint(dict(response.headers))
```

```python
todos = json.loads(response.text)
```

```python
 todos[:10]
```

```python
# Map of userId to number of complete TODOs for that user
todos_by_user = {}

# Increment complete TODOs count for each user.
for todo in todos:
    if todo["completed"]:
        try:
            # Increment the existing user's count.
            todos_by_user[todo["userId"]] += 1
        except KeyError:
            # This user has not been seen. Set their count to 1.
            todos_by_user[todo["userId"]] = 1
```

```python
todos_by_user
```

```python
todos_by_user.items()
```

```python
list(todos_by_user.items())
```

```python
# Create a sorted list of (userId, num_complete) pairs.
top_users = sorted(todos_by_user.items(), 
                   key=lambda x: x[1], reverse=True)
```

```python
# Get the maximum number of complete TODOs.
max_complete = top_users[0][1]
```

```python
# Create a list of all users who have completed
# the maximum number of TODOs.
users = []
for user, num_complete in top_users:
    if num_complete < max_complete:
        break
    users.append(user)
print(users)
```

```python
max_users = " and ".join([str(s) for s in users])
```

```python
s = "s" if len(users) > 1 else ""
```

```python
print(f"user{s} {max_users} completed {max_complete} TODOs")
```

```python
completed_todos = {}
for i in users:
    completed_todos[i]=[]
completed_todos
```

```python
for todo in todos:
    if todo["completed"]:
        if todo["userId"] in completed_todos:
            completed_todos[todo["userId"]].append(todo["id"])
completed_todos
```

```python
fn = 'json_top_2.txt'
with open(fn, 'w') as fid:
    json.dump(completed_todos, fid)
```

## Posting and Getting Examples

```python
import requests
from util_functions import values

page = 'http://localhost/cgi-bin/form01.py'
query = {'first': 'joe', 'second': 'banks'}
data = {"a_key": "a_value"}

method = 'post'

if method == 'get':
    response_obj  = requests.get(page, params=query)
elif method == 'post':
    response_obj  = requests.post(page, data=query)
else:
    response_obj  = None

print(response_obj)
values(response_obj)
```

```python
request_obj = response_obj.request
print(request_obj.method)
print(request_obj.headers)
print(request_obj.body)
```

```python
values(request_obj)
```
